```python
records = [{"id": 1, "name": "Anna", "age": "NA"},
    {"id": 2, "name": "", "age": 25},
    {"id": 3, "name": "John", "age": "null"} ]
print(records)

def null_handling(rows):
    res = []
    for i in records:
        d={}
        for k,v in i.items():
            d[k] = None if v in ("NA", "null", "") else v
        res.append(d)
    return res

null_handling(records)
```

```python
rows = [
    (1, "A", "a@gmail.com"),
    (2, "B", "b@gmail.com"),
    (3, "C", "a@gmail.com")
]

def dedup_by_email(rows):
    seen = set()
    res = []
    for r in rows:
        if r[2] not in seen:
            seen.add(r[2])
            res.append(r)
    return res

dedup_by_email(rows)
```

```python
data = ["10", "10.5", 20, "abc", None]

def clean_numbers(arr):
    out = []
    for x in arr:
        try:
            out.append(int(float(x)))
        except:
            pass
    return out

clean_numbers(data)
```

```python
csv_data = """id,name,age
1, Alice , 20
2,Bob,25
"""

import csv
from io import StringIO

def load_csv(data):
    f = StringIO(data)
    reader = csv.DictReader(f)
    return [{k.strip(): v.strip() for k, v in row.items()} for row in reader]

load_csv(csv_data)
```

```python
data = [1,2,3,4,5]
k = 3

def batch_data(data, k):
    return [data[i:i+k] for i in range(0, len(data), k)]

batch_data(data, k)

print(k)
```
